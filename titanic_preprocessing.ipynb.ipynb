```python
df['Age'] = df['Age'].fillna(df['Age'].median())    # Fill missing Age values with median
```

```python
df["Cabin"].nunique() ,df["Cabin"].unique()       # Shows number of unique Cabin values and the values themselves
```

```python
df.drop('Cabin', axis=1, inplace=True)       # Drop the Cabin column
```

```python
df["Embarked"].nunique() ,df["Embarked"].unique()      # Shows number of unique values and the values in Embarked
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])    # Fill missing Embarked values with mode
```

```python
df.isnull().sum()     # Check if any missing values remain after filling
```

```python
df.select_dtypes(include='object')     # Select columns with object (categorical/text) data type
```

```python
cat_columns = df.select_dtypes(include='object').columns
cat_columns                                                     # Display categorical column names
```

```python
df.drop('Name', axis=1, inplace=True)   # Drop the Name column from the dataset (not useful for analysis/model).
```

```python
df.drop('Ticket', axis=1, inplace=True)    # Drop the Ticket column from the dataset (not useful for analysis/model).
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})    #Encode the **Sex** column into numeric values (male = 0, female = 1).
```

```python
df = pd.get_dummies(df, columns=['Embarked'], dtype=int)  # Convert the Embarked categorical column into one-hot encoded columns (0/1).
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import StandardScaler     # Import StandardScaler to scale numeric features.
scaler = StandardScaler()    # Create a scaler object.

df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])    # Standardize the Age and Fare columns (mean = 0, std = 1).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Age'])
plt.show()

sns.boxplot(x=df['Fare'])
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[(df['Fare'] >= lower) & (df['Fare'] <= upper)]
```
